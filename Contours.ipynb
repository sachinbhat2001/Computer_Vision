```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('square.png') #importing the image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)#converting image from BGR color channel to RGB color channel
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)#converting colored image into gray scale
```

```python
#applying binary threshold and displaying the image
retval,binary = cv2.threshold(gray, 210,225, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray')
```

```python
#finding contours and displaying the contour list
cnts = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(cnts)
```

```python
#this is not a mandatory step
#this is used to make a copy of orginal image just for future reference
contours_image = np.copy(image)
contours = cnts[0] if len(cnts) == 2 else cnts[1]
```

```python
#Drawing the contour for the image and displaying it. 
contours_image = cv2.drawContours(contours_image, contours, -1, (0,0,0), 3)
plt.imshow(contours_image)
```
